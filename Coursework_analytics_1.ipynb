```python
import datetime
from datetime import datetime, timedelta
import dateutil

def age(date): 
    try:
        # Переводим дату рождения в нужный формат
        date = datetime.strptime(date,'%Y-%m-%d')
        # Вычисляем текущую дату
        now = datetime.utcnow() 
        now = now.date() 
        # Находим разницу между текущей датой и датой рождения.
        age = dateutil.relativedelta.relativedelta(now, date) 
        age = age.years 
        # Исключим часть ошибочных значений возраста, поставив условие, 
        # что возраст не должен быть меньше 7, а также больше 100
        if age < 7:
            age = None
        elif age > 100:
            age = None
        else:
            return age
    except TypeError:
        pass
```

```python
def convert_to_year_month_s (row):
    new_date = pd.to_datetime(row['start_date'])
    year_month = datetime.strftime(new_date,'%Y-%m')
    year_month = datetime.strptime(year_month, '%Y-%m')
    return year_month
```

```python
df_first_lesson = total_df[(total_df['module_number']==1)&(total_df['is_homework']==True)]
df_first_lesson['start_date_datetime']=df_first_lesson.apply(convert_to_year_month_s, axis=1)
df_first_lesson = df_first_lesson[['start_date_datetime','course_title', 'student_id'
                                  ]].groupby(['start_date_datetime','course_title']).count().reset_index()
df_first_lesson = df_first_lesson.pivot(index = 'start_date_datetime', columns = 'course_title'
                                       ).fillna(0).reset_index()
```
